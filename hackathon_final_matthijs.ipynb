```python
import pandas as pd

# Controleer de kolomnamen in de DataFrame
print(excel.columns)

# Veronderstel dat je kolommen 0 t/m 23 numeriek zijn, zonder aanhalingstekens
hour_columns = list(range(24))  # Maak een lijst met de kolomnummers (0 tot en met 23)

# Zorg ervoor dat de 'Datum' kolom naar datetime wordt omgezet
excel['Datum'] = pd.to_datetime(excel['Datum'])

# Gebruik de pandas melt-functie om de uren (kolommen 0-23) naar rijen om te zetten
df_melted = pd.melt(excel, id_vars=['Datum'], value_vars=hour_columns,
                    var_name='Uur', value_name='Waarde')

# Combineer de 'Datum' kolom met de uur-kolom om een datetime-object te maken
df_melted['Datum'] = df_melted['Datum'] + pd.to_timedelta(df_melted['Uur'].astype(int), unit='h')

# Verwijder de overbodige 'Uur'-kolom
df_melted = df_melted.drop(columns=['Uur'])

# Sorteer de DataFrame op datum
df_melted = df_melted.sort_values('Datum').reset_index(drop=True)

# Toon het resultaat
df_melted.head()
```
